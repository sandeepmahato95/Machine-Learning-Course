```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Initialize an empty list to store data from all pages
all_data = []

for i in range(1, 25):
    url = 'https://www.scrapethissite.com/pages/forms/?page_num={}'.format(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html')

    # Extract table headers only once (they are the same on all pages)
    if i == 1:
        table_titles = soup.find_all('th')
        table_titles_names = [titles.text.strip() for titles in table_titles]

    # Find the table on the current page
    table = soup.find('table')

    # Extract data rows from the table
    row_number = table.find_all('tr')
    for row_no in row_number[1:]:  # Skip the header row
        row_data = row_no.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        all_data.append(individual_row_data)

# Create the DataFrame outside the loop using the collected data
df = pd.DataFrame(all_data, columns=table_titles_names)
df
```

```python
# Save the dataframe to a CSV file named 'df.csv'
df.to_csv('NHL Hockey Teams data.csv', index=False) # index=False prevents writing row indices to the file
```
