# 14. Convertors

```python
pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
def rename(name):
    if name=='Kolkata Knight Riders':
        return 'KKR'
    elif name=='Royal Challengers Bangalore':
        return 'RCB'
    elif name=='Kings XI Punjab':
        return 'KXIP'
    elif name=='Mumbai Indians':
        return 'MI'
    elif name=='Chennai Super Kings':
        return 'CSK'
    elif name=='Rajasthan Royals':
        return 'RR'   
    else:
        return name
    
```

```python
rename('Kolkata Knight Riders')
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename,'team2':rename, 'toss_winner':rename, 'winner':rename})

# To apply transformation on any column values
```

# 17. To Skip bad lines using <span style="color:red">*on_bad_lines*</span>

```python
pd.read_csv('BX-Books.csv')
```

BX-Books.csv file is separated by semi colon (;) and encoded (latin-1)

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1")
```

**on_bad_lines{‘error’, ‘warn’, ‘skip’}** or Callable, default ‘error’
Specifies what to do upon encountering a bad line (a line with too many fields). Allowed values are :

**error**, raise an Exception when a bad line is encountered.

**warn**, raise a warning when a bad line is encountered and skip that line.

**skip**, skip bad lines without raising or warning when they are encountered

[Reference link](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

and few rows have more values than the column

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines='warn')
```

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines='skip')
```

**low_memory**: bool, default True

Internally process the file in chunks, resulting in lower memory use while parsing, but _possibly mixed type inference_. To ensure no mixed types either set False, or specify the type with the dtype parameter. 

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines='skip',low_memory=False)
```

headers names are not correct.

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines='skip',low_memory=False).info()
```

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines='skip',low_memory=False, 
            names=['ISBN','Book-Title','Book-Author','Year of Publication', 'Publisher', 'Image URL'])
```

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines='skip',low_memory=False, 
            names=['ISBN','Book-Title','Book-Author','Year of Publication', 'Publisher', 'Image URL']).info()
```
