Connecting to SQL Database using SQLAlchemy in Python

The **create_engine()** method of sqlalchemy library takes in the connection URL and returns a sqlalchemy engine that references both a Dialect and a Pool, which together interpret the DBAPI’s module functions as well as the behavior of the database

Reference:

https://www.geeksforgeeks.org/connecting-to-sql-database-using-sqlalchemy-in-python/?ref=ml_lbp

https://docs.sqlalchemy.org/en/20/core/engines.html

```python
# Importing The SQAlchemy library's 'create_engine' method

from sqlalchemy import create_engine
```

The connection URL to the database of type “dialect+driver://username:password@host:port/database”.

Format of connection string is:

mysql+pymysql://username:password@host:port/database_name

```python
engine = create_engine("mysql+pymysql://root:123@localhost:3306/world")

#engine = create_engine("mysql+pymysql://root:123@localhost/world")
```

<h3 style="color:black">Read entire table in a dataframe using <span style="color:red">read_sql_table</span></h3>

```python
df1 = pd.read_sql_table('city',engine)
df1
```

```python
query='SELECT * FROM city'
df2= pd.read_sql(query,engine)
df2
```

```python
query='SELECT * FROM countrylanguage'
df3= pd.read_sql_query(query,engine)
df3
```

<h3 style="color:black"><span style="color:red">read_sql</span> is a wrapper around read_sql_query and read_sql_table</h3>

```python
df4= pd.read_sql('country',engine)
df4
```
