## DataFrame Information

The dataset contains 96,845 entries, each with 8 different columns of information. Here is a brief overview of what each column represents:

1. **Year**: This column indicates the year when the data was recorded. It contains numerical values (floating-point numbers).
2. **Month_Name**: This column specifies the name of the month for each entry. It contains textual data.
3. **Date**: This column provides the exact date for each entry. It contains textual data.
4. **State**: This column indicates the state where the data was recorded. It contains textual data.
5. **Vehicle_Class**: This column categorizes the type of vehicle based on its class. It contains textual data.
6. **Vehicle_Category**: This column further categorizes the vehicle. It contains textual data.
7. **Vehicle_Type**: This column specifies the type of vehicle. It contains textual data.
8. **EV_Sales_Quantity**: This column shows the quantity of electric vehicle (EV) sales. It contains numerical values (floating-point numbers).

## Skewness for the Year Column

The skewness value for the 'Year' column is approximately -0.0023. Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean.

- **Observation**: The skewness value is very close to zero, which indicates that the distribution of the 'Year' data is nearly **symmetrical**, which means that the data is evenly distributed around the mean, with no significant tail on either side. The years in the dataset are fairly balanced and do not lean heavily towards earlier or later years.


## Box Plot for Years

The box plot above visualizes the distribution of years in the dataset.

- **Box**: The central box represents the middle 50% of the data, known as the interquartile range (IQR). This is where the majority of the data points lie.
- **Line Inside the Box**: The line inside the box represents the median (or the 50th percentile) of the data. In this plot, the median year is around 2018.
- **Whiskers**: The lines extending from the box (whiskers) show the range of the data, excluding outliers. They typically extend to 1.5 times the IQR from the edges of the box.
- **Outliers**: Any data points that fall outside the whiskers are considered outliers. In this plot, there are no outliers visible.

From this box plot, we can see that the majority of the data points fall between approximately 2016 and 2020, with the median year being around 2018. There are no extreme outliers in the dataset.

#### The code to convert `Month_Name` (jan to dec) to `Month_Number` (1 to 12)

from time import strptime

Assuming your DataFrame is named 'df' and the month column is named 'Month_Name' 

df['Month_Number'] = [strptime(str(x), '%b').tm_mon for x in df['Month_Name']]
